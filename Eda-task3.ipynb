# **ðŸ”¤ Section 3: Encoding Categorical Variables**

15. Which columns need encoding before applying machine learning models? Why?

Columns that need encoding: person_gender, person_education, person_home_ownership, loan_intent, previous_loan_defaults_on_file
Because these columns are categorical (contain text values), and machine learning models require numerical input.

```python
df['person_gender'] = df['person_gender'].map({'male': 1, 'female': 0})
df = pd.get_dummies(df, columns=['person_home_ownership'], drop_first=True, dtype='int')
df = pd.get_dummies(df, columns=['loan_intent'], drop_first=True, dtype='int')
```

```python
edu_lvl = {
    'High School': 0,
    'Associate': 1,
    'Bachelor': 2,
    'Master': 3,
    'Doctorate': 4
}

df['person_education'] = df['person_education'].replace(edu_lvl)
df
```

22. Which features have vastly different ranges and may require scaling?

23. Apply MinMaxScaler or StandardScaler to relevant numerical features. Show
before and after scaling plots.

```python
num_features = ['person_age', 'person_income', 'loan_amnt', 'loan_int_rate',
                'loan_percent_income', 'cb_person_cred_hist_length',
                'credit_score', 'person_emp_exp']

plt.figure(figsize=(16, 8))

sns.histplot(df['person_age'], kde=True, ax=plt.subplot(2, 4, 1))
plt.title('Original person_age')

sns.histplot(df['person_income'], kde=True, ax=plt.subplot(2, 4, 2))
plt.title('Original person_income')

sns.histplot(df['loan_amnt'], kde=True, ax=plt.subplot(2, 4, 3))
plt.title('Original loan_amnt')

sns.histplot(df['loan_int_rate'], kde=True, ax=plt.subplot(2, 4, 4))
plt.title('Original loan_int_rate')

sns.histplot(df['loan_percent_income'], kde=True, ax=plt.subplot(2, 4, 5))
plt.title('Original loan_percent_income')

sns.histplot(df['cb_person_cred_hist_length'], kde=True, ax=plt.subplot(2, 4, 6))
plt.title('Original cb_person_cred_hist_length')

sns.histplot(df['credit_score'], kde=True, ax=plt.subplot(2, 4, 7))
plt.title('Original credit_score')

sns.histplot(df['person_emp_exp'], kde=True, ax=plt.subplot(2, 4, 8))
plt.title('Original person_emp_exp')

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_standard = df.copy()
df_standard[num_features] = scaler.fit_transform(df_standard[num_features])
```

```python
plt.figure(figsize=(16, 8))

plt.subplot(2, 4, 1)
sns.histplot(df_standard['person_age'], kde=True)
plt.title('Standard Scaled person_age')

plt.subplot(2, 4, 2)
sns.histplot(df_standard['person_income'], kde=True)
plt.title('Standard Scaled person_income')

plt.subplot(2, 4, 3)
sns.histplot(df_standard['loan_amnt'], kde=True)
plt.title('Standard Scaled loan_amnt')

plt.subplot(2, 4, 4)
sns.histplot(df_standard['loan_int_rate'], kde=True)
plt.title('Standard Scaled loan_int_rate')

plt.subplot(2, 4, 5)
sns.histplot(df_standard['loan_percent_income'], kde=True)
plt.title('Standard Scaled loan_percent_income')

plt.subplot(2, 4, 6)
sns.histplot(df_standard['cb_person_cred_hist_length'], kde=True)
plt.title('Standard Scaled cb_person_cred_hist_length')

plt.subplot(2, 4, 7)
sns.histplot(df_standard['credit_score'], kde=True)
plt.title('Standard Scaled credit_score')

plt.subplot(2, 4, 8)
sns.histplot(df_standard['person_emp_exp'], kde=True)
plt.title('Standard Scaled person_emp_exp')

plt.tight_layout()
plt.show()
```
